```python
import os, json, pickle
```

```python
datafile = os.path.join('../data_dir/doordash_categorized.pkl')
print(datafile)
with open(datafile, 'rb') as f:
    json = pickle.load(f)
```

```python
json[0]
```

```python
target_sections = [
    'sushi', 'pizza', 'soup', 'salad', 'sandwich', 'burger'
]
```

```python
from collections import defaultdict
sec_to_word = defaultdict(set)
```

```python
from collections import Counter
sec_counter = Counter()
for entry in json:
    title = entry['title']
    label = entry['labels']['category']
    section = entry['metadata']['section']
    sec_counter[section] += 1
    san_section = section.lower()
    for t in target_sections:
        if t in san_section:
            sec_to_word[t].add(san_section)
```

```python
sec_counter.most_common()
```
