```python
tmux_mins = 0
other_mins = 0

for line in lines:
    arr = line.split("  ")
    arr = [a for a in arr if a]
    if arr[0].startswith('uma'):
        print(arr)
        tmux = 'tmux' in arr[1]
        if tmux:
            print('tmux')
        time_entry = arr[-1]
        if "(" not in time_entry:
            assert 'still logged in' in time_entry
            continue  # still logged in
        time = time_entry[time_entry.find("(")+1:time_entry.find(")")]
        mins = 0
        if "+" in time:
            day, rest = time.split("+")
            mins += int(day) * 24 * 60
        else:
            rest = time
        h, m = rest.split(":")
        mins += int(h) * 60 + int(m)
        print(mins)
        
        if tmux:
            tmux_mins += mins
        else:
            other_mins += mins
```
